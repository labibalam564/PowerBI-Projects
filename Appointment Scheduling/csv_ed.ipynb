# Created to fix the 'Client ID' value in row index 25 as it was blank. Additionally, this python code was used to remove irrelevant data from the dataset to improve the overall performance of the ETL process on Power Query Editor.

```python
import pandas as pd
from pandas import DataFrame
```

```python
df = pd.read_csv('data_file.csv')
df.loc[24, 'Client ID'] = 25
df.drop(['First Name', 'Last Name', 'Waiting Time'], inplace=True, axis=1)
df.to_csv('data_file.csv', index=False)
```

```python
df['Requesting date'] = pd.to_datetime(df['Requesting date'])
df['Appointment date'] = pd.to_datetime(df['Appointment date'])

df['request_date'] = df['Requesting date'].dt.strftime('%m/%d/%Y')
df['appointment_date'] = df['Appointment date'].dt.strftime('%m/%d/%Y')

df.head()
```

```python
df.drop(['Requesting date', 'Appointment date'], inplace=True, axis=1)
df.head()
```

```python
df.to_csv('data_file.csv', index=False)
```
